<h1>DATA CLEANSING</h1>

```python
accident.isnull().sum()
```

```python
accident['Latitude'] = accident['Latitude'].fillna(accident['Latitude'].mode()[0])
accident['Longitude'] = accident['Longitude'].fillna(accident['Longitude'].mode()[0])
accident['Road_Surface_Conditions'] = accident['Road_Surface_Conditions'].fillna('unaccounted')
accident['Road_Type'] = accident['Road_Type'].fillna('unaccounted')
accident['Weather_Conditions'] = accident['Weather_Conditions'].fillna('unaccounted')
accident['Urban_or_Rural_Area'] = accident['Urban_or_Rural_Area'].fillna('unaccounted')
```

```python
accident.isnull().sum()
```

```python
accident['Accident_Severity'] = accident['Accident_Severity'].astype('category')
```

<h1>INSIGHT 18</h1>

```python

# Create a pie chart
plt.figure(figsize=(8, 8))
weekend_weekday_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])

# Add title
plt.title('Weekend vs Weekday Accidents')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```

```python
# Assuming 'weekend_weekday_counts' is your Series
# Convert the Series to a DataFrame for easier plotting
weekend_weekday_df = weekend_weekday_counts.reset_index()
weekend_weekday_df.columns = ['Day_Type', 'Accident_Count']  # Rename columns for clarity

# Sort the DataFrame by 'Day_Type' for consistent ordering
weekend_weekday_df['Day_Type'] = pd.Categorical(weekend_weekday_df['Day_Type'], categories=['Weekday', 'Weekend'], ordered=True)
weekend_weekday_df = weekend_weekday_df.sort_values('Day_Type')

# Plotting the bar graph
plt.figure(figsize=(8, 6))  # Set the figure size
sns.barplot(data=weekend_weekday_df, x='Day_Type', y='Accident_Count', palette='viridis')

# Add labels and title
plt.title("Accidents on Weekdays vs Weekends", fontsize=16)
plt.xlabel("Day Type", fontsize=12)
plt.ylabel("Number of Accidents", fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

<H3>-The result shows that weekdays are associated with higher accident rates compared to weekends.</H3>

<h1>INSIGHT 19</h1>
